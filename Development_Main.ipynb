<h2 align=center>Function Declarations</h2>

```python
def read_file(filename):
    with open(filename, 'r') as file:
        file_content = file.read()
    return file_content

def create_gtts_object(text_to_read, language, tld):
    gtts_object = gTTS(text=text_to_read,
                      lang=language,
                      slow=False,
                      lang_check=True,
                      tld=tld)
    
    # Save the gTTS object as a WAV file
    file_path = "files/gtts.wav"
    gtts_object.save(file_path)
    
    # Get file size
    file_size = os.path.getsize(file_path) / (1024 * 1024)  # Convert bytes to megabytes
    
    # Print file size
    print(f"File size: {file_size:.3f} MB")
    
    # Read the file with progress bar
    with tqdm(total=file_size, unit="MB", unit_scale=True) as pbar:
        with open(file_path, 'rb') as file:
            while True:
                chunk = file.read(1024)
                if not chunk:
                    break
                pbar.update(len(chunk) / (1024 * 1024))  # Update with chunk size in megabytes
    
    return gtts_object
```
